### Platform Efficiency

```python
eff = df.groupby('Primary_Search_Platform')[['Offer_Received','Offer_Salary','Time_to_Offer_Days','Applications_Submitted']].mean().reset_index().round(2)
eff
```

## 1. Which platform helps a candidate to receive an offer the quickest?
There is a need to make students aware of the most effective job search platforms so they are able to optimise their time in job searching

```python
time_to_offer = eff[['Primary_Search_Platform','Time_to_Offer_Days']].sort_values(by="Time_to_Offer_Days")
time_to_offer
```

```python
sns.barplot(x=time_to_offer.Primary_Search_Platform,y=time_to_offer.Time_to_Offer_Days,hue=time_to_offer.Primary_Search_Platform)
plt.title('Time Taken to Receive Offer')
```

I created a platform efficacy table and from it I was able to select the two features that I needed for this analysis which are 'Primary_Search_Platform' and 'Time_to_Offer_Days'. I then sorted that dataframe by the 'Time_to_Offer_Days' column. I then created a barplot. This bar plot clearly indicates that it takes about 28 days to get an offer on Indeed while it might take you close to 2 months if you apply using Linkedin.

## 2. Which Programme Commands The Highest Salary?
For a school like Gisma that focuses mainly on STEM and Business oriented programs, there is a need to inform students of how competitive Gisma courses are and the reason why you should opt for such courses as compared to others.

## 3. Do Internships Influence Job Offers?
A student would want to know if Internships are that necessary

```python
int_ship = df.groupby('Prior_Internships')['Offer_Received'].mean()
int_ship
```

```python
sns.lineplot(x=int_ship.index,y=int_ship.values, marker='o',color='purple')
plt.title('Job Offers Probability According to Number of Internships')
```

I took the average of the percentages of offers received according to the number of prior internships an individual did and created a line graph for visual representation and easier interpretation.
We can see from the tragectory that a person's chance of getting a job offer increases with more internships

## 4. Does A Student's GPA influence The Possibility of Receiving A Job Offer?
It is necessary to explain to a student why they need to take their studies seriously.

```python
df['GPA_Group']=pd.cut(df['GPA'], bins=[2.0,2.5,3.0,3.5,4.0])
gpa = df.groupby('GPA_Group')['Offer_Received'].mean()
gpa
```

```python
sns.barplot(x=gpa.index.astype(str),y=gpa.values,legend=False, palette='coolwarm',hue=gpa.index.astype(str))
plt.title('Probabilty of Job Offers per GPA')
plt.xlabel('GPA')
plt.ylabel('Probabilty of Job Offer')
```

To be able to calculate the probability of job offers per a persons GPA, I had to group all GPAs that fall within a certain bracket. The unique values for GPA are so many so it is only possible to fit a few figures on the graph that is why there was a need to group them. I then found the average offers received by each group. The graph indicates that the higher the GPA, the higher the chances of getting a job offer. These results will serve as both encouragement and advise to students
